```python
import pandas as pd

# Replace 'your_dataset.csv' with the actual file path
# client_0_bills = pd.read(csv/content/drive/MyDrive/STEG_BILLING_HISTORY.csv.crdownload)
client_0_bills = pd.read_csv('STEG_BILLING_HISTORY.csv')

# Display the first few rows of the dataset
client_0_bills.head()


```

```python
# Number of rows and columns
num_rows, num_cols = client_0_bills.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
#How many categorical features are in the dataset
num_categorical = len(client_0_bills.select_dtypes(include='object').columns)
print("Number of categorical features:", num_categorical)
```

```python
#How much memory space does the dataset consume
memory_usage = client_0_bills.memory_usage().sum() / 1024**2  # Assuming average row size
print("Estimated memory usage:", round(memory_usage, 2), "MB")
```

```python
#Inspect the dataset for potential missing values
missing_values = client_0_bills.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
#Select your strategy to handle missing values
#Data dropping
#Data dropping using the dropna() function is the easiest way  to remove observations or features with missing values from the dataframe.
drop_na_strategy = client_0_bills.dropna()
drop_na_strategy.info()
```

```python
#Run a descriptive analysis on numeric features(columns)
print(client_0_bills.describe())
```

```python

#Selecting bills for client with id 'train_Client_0'
# Method 1:
client_records_loc = client_0_bills.loc[client_0_bills['client_id'] == 'train_Client_0']

# Method 2:
client_records_boolean = client_0_bills[client_0_bills['client_id'] == 'train_Client_0']

# Display the results from both methods
print("Using .loc[]:")
print(client_records_loc)
print("\nUsing boolean indexing:")
print(client_records_boolean)
```

```python
#Transform the counter_type feature to a numeric value using an encoder of your choice
from sklearn.preprocessing import LabelEncoder

#Initializing the encoder
encoder = LabelEncoder()

#Fitting and transforming the 'counter_type' column
client_0_bills['counter_type'] = encoder.fit_transform(client_0_bills['counter_type'])

#Displaying the transformed column
print("\nTransformed 'counter_type':")
print(client_0_bills['counter_type'])
```

```python
# Delete the 'counter_statue' feature
client_0_bills.drop(columns=['counter_statue'], inplace=True)

# Display the DataFrame after deletion to confirm
print("\nDataFrame after deleting 'counter_statue':")
print(client_0_bills.head())
```
